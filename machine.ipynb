```python
df = df.sort_values(['product_name', 'size', 'snapshot_date'])

sku_cols = ['product_name', 'size']
```

```python
# convert snapshot_date to datetime and extract year/month/day
df['snapshot_date'] = pd.to_datetime(df['snapshot_date'])
df['s_year'] = df['snapshot_date'].dt.year
df['s_month'] = df['snapshot_date'].dt.month

df.head()
```

```python
def create_features(data):

    data = data.sort_values(['product_name', 'size', 'snapshot_date'])

    sku_cols = ['product_name', 'size']

    # Lag features
    for lag in [1, 2, 3]:
        data[f'sales_lag_{lag}'] = (
            data.groupby(sku_cols)['total_sold'].shift(lag)
        )

    # Rolling mean
    data['rolling_mean_3'] = (
        data.groupby(sku_cols)['total_sold']
            .transform(lambda x: x.shift(1).rolling(3).mean())
    )

    # Rolling std
    data['rolling_std_3'] = (
        data.groupby(sku_cols)['total_sold']
            .transform(lambda x: x.shift(1).rolling(3).std())
    )

    # Seasonality
    data['month_sin'] = np.sin(2 * np.pi * data['s_month'] / 12)
    data['month_cos'] = np.cos(2 * np.pi * data['s_month'] / 12)

    # Target (next month sales)
    data['target'] = (
        data.groupby(sku_cols)['total_sold'].shift(-1)
    )

    return data

df_feat = create_features(df)
df_feat = df_feat.dropna()
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures, StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import VotingRegressor
from sklearn.compose import ColumnTransformer
from sklearn.metrics import mean_absolute_error

# -----------------------------
# Build Voting Ensemble
# -----------------------------
def build_model(numeric_features, categorical_features):
    # Preprocessing
    numeric_transform = Pipeline([
        ("scaler", StandardScaler()),
        ("poly", PolynomialFeatures(degree=2, include_bias=False))
    ])
    categorical_transform = OneHotEncoder(handle_unknown='ignore')
    preprocessor = ColumnTransformer([
        ("num", numeric_transform, numeric_features),
        ("cat", categorical_transform, categorical_features)
    ])

    # Base models
    poly_model = Pipeline([
        ("pre", preprocessor),
        ("linreg", LinearRegression())
    ])

    knn_model = Pipeline([
        ("pre", preprocessor),
        ("knn", KNeighborsRegressor(n_neighbors=3))
    ])

    tree_model = Pipeline([
        ("pre", preprocessor),
        ("tree", DecisionTreeRegressor(max_depth=3, random_state=42))
    ])

    model = VotingRegressor([
        ("poly", poly_model),
        ("knn", knn_model),
        ("tree", tree_model)
    ])

    return model

# -----------------------------
# Prepare features
# -----------------------------
df_feat = df.copy()

# Lag features: only previous month sales (since history is short)
df_feat['sales_lag_1'] = df_feat.groupby(['product_name','size'])['total_sold'].shift(1)

# Target: next month sales
df_feat['target'] = df_feat.groupby(['product_name','size'])['total_sold'].shift(-1)

# Drop rows with NaN (first/last month per SKU)
df_feat = df_feat.dropna().reset_index(drop=True)

numeric_features = ['total_sold', 'quantity_on_hand', 'sales_lag_1']
categorical_features = ['product_name', 'size']

# -----------------------------
# Train global model
# -----------------------------
model = build_model(numeric_features, categorical_features)

X = df_feat[numeric_features + categorical_features]
y = df_feat['target']

model.fit(X, y)

# Predict next month sales
df_feat['pred_demand'] = model.predict(X)
df_feat['pred_demand'] = df_feat['pred_demand'].clip(lower=0)

# -----------------------------
# Compute reorder using base-stock rule
# -----------------------------
z = 1.28  # safety factor
holding_cost = 1
penalty_cost = 5

# Standard deviation per SKU
sku_std = df_feat.groupby(['product_name','size'])['total_sold'].transform('std').fillna(0)

df_feat['base_stock'] = df_feat['pred_demand'] + z * sku_std

# Inventory simulation
df_feat['inventory'] = df_feat['quantity_on_hand']
df_feat['reorder_qty'] = np.maximum(0, df_feat['base_stock'] - df_feat['inventory'])
df_feat['inventory'] += df_feat['reorder_qty']

df_feat['sales'] = df_feat[['inventory','target']].min(axis=1)
df_feat['lost_sales'] = np.maximum(0, df_feat['target'] - df_feat['inventory'])
df_feat['inventory'] -= df_feat['sales']

df_feat['holding_cost'] = df_feat['inventory'] * holding_cost
df_feat['penalty_cost'] = df_feat['lost_sales'] * penalty_cost
df_feat['total_cost'] = df_feat['holding_cost'] + df_feat['penalty_cost']

# -----------------------------
# Metrics
# -----------------------------
mae = mean_absolute_error(df_feat['target'], df_feat['pred_demand'])
total_cost = df_feat['total_cost'].sum()
stockout_rate = (df_feat['lost_sales'] > 0).mean()

print("MAE:", mae)
print("Total Cost:", total_cost)
print("Stockout Rate:", stockout_rate)

# View result
df_feat[['product_name','size','snapshot_date','target','pred_demand','reorder_qty','inventory','lost_sales','total_cost']]
```
